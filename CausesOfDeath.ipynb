We always get 244 as a count for null counts. Maybe there is some specific country that we cannot get data from ?

```python
death_columns = [col for col in df.columns if col.startswith('Deaths')]
```

```python
# Group by 'Entity' and check for NaNs in 'Deaths' columns
grouped = df.groupby('Entity')[death_columns].apply(lambda x: x.isnull().any(axis=1).sum())

# Convert the resulting Series to a DataFrame for better presentation if needed
count_nan_per_entity = grouped.reset_index(name='Count of NaNs')

# Display the count
print("Number of rows with at least one NaN per entity:\n", count_nan_per_entity)
```

We looked whether there is any specific missing values that would make sense to fill with anything other than 0, like forward filling or mean and so on. But 244/8k is okay for our analysis to fill with 0. Nonetheless, we will fill with mean to approach conservatively.

```python
# Identify numeric columns
numeric_columns = df.select_dtypes(include=['number']).columns

# Fill NaN values in the numeric columns with the mean of the respective country group
df[numeric_columns] = df.groupby('Entity')[numeric_columns].transform(lambda x: x.fillna(x.mean()))
```

```python
# Ensure all death count columns are numeric
death_columns = df.columns[3:]  # All columns except 'Entity', 'Code', 'Year'
df[death_columns] = df[death_columns].apply(pd.to_numeric, errors='coerce')

# Melt the dataframe to long format
melted_df = df.melt(id_vars=['Entity', 'Code', 'Year'], 
                    var_name='Cause', 
                    value_name='Death_Count')

# Group by Year and Cause, then sum the Death_Count
yearly_deaths = melted_df.groupby(['Year', 'Cause'])['Death_Count'].sum().reset_index()

# Find the leading cause of death for each year
leading_causes = yearly_deaths.loc[yearly_deaths.groupby('Year')['Death_Count'].idxmax()]

# Merge with world population data
leading_causes = leading_causes.merge(world_population, on='Year')

# Calculate death rate per capita
leading_causes['Death_Rate'] = leading_causes['Death_Count'] / (leading_causes['Population'] * 1e9)  # converting billion to actual population

# Plotting the leading causes of death over the years per capita
plt.figure(figsize=(20, 12))
plt.plot(leading_causes['Year'], leading_causes['Death_Rate'], marker='o')

# Adding labels for each point
for i, row in leading_causes.iterrows():
    plt.text(row['Year'], row['Death_Rate'], row['Cause'], fontsize=12, ha='right')

plt.title('Leading Cause of Death Per Capita Over the Years')
plt.xlabel('Year')
plt.ylabel('Death Rate (per person)')
plt.grid(True)
plt.show()
```

### I am curious about Germany since I live here:

```python
# Filter the data for Germany
germany_data = df[df['Entity'] == 'Germany']
```

```python
# Melt the dataframe to long format
melted_germany = germany_data.melt(id_vars=['Entity', 'Code', 'Year'], 
                                   var_name='Cause', 
                                   value_name='Death_Count')

# Ensure Death_Count is numeric
melted_germany['Death_Count'] = pd.to_numeric(melted_germany['Death_Count'], errors='coerce')

# Group by Year and Cause, then sum the Death_Count
yearly_deaths_germany = melted_germany.groupby(['Year', 'Cause'])['Death_Count'].sum().reset_index()

# Find top 5 leading causes of death for each year
top5_deaths_germany = yearly_deaths_germany.groupby('Year').apply(
    lambda x: x.nlargest(5, 'Death_Count')
).reset_index(drop=True)

top5_deaths_germany
```

```python
# Frequency of causes appearing in top 5
top5_frequency = top5_deaths_germany['Cause'].value_counts()
top5_frequency
```

```python
# Plotting trends for specific causes of death over the years
causes_of_interest = top5_frequency.index[:5]  # Top 5 frequent causes

plt.figure(figsize=(14, 8))
for cause in causes_of_interest:
    cause_data = yearly_deaths_germany[yearly_deaths_germany['Cause'] == cause]
    plt.plot(cause_data['Year'], cause_data['Death_Count'], marker='o', label=cause)

plt.title('Trends of Leading Causes of Death in Germany Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Deaths')
plt.legend()
plt.grid(True)
plt.show()
```
